```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("global_power_plant_database.csv")
```

```python
df["country"].value_counts()
```

```python
df_all = df.drop(['country_long', 'name', 'url', 'geolocation_source', 'wepp_id'], axis=1)
```

```python
df_all['gppd_idnr'].is_unique
```

```python
df_all["primary_fuel_num"] = df_all["primary_fuel"].map({'NaN':0, 'Hydro':1, 'Solar':2, 'Wind':3, 'Gas':4, 'Coal':5, 'Oil':6, 'Biomass':7, 'Waste':8, 'Nuclear':9, 'Geothermal':10, 'Storage':11, 'Other':12, 'Cogeneration':13, 'Petcoke':14, 'Wave and Tidal':15})
df_all["other_fuel1_num"] = df_all["other_fuel1"].map({'NaN':0, 'Hydro':1, 'Solar':2, 'Wind':3, 'Gas':4, 'Coal':5, 'Oil':6, 'Biomass':7, 'Waste':8, 'Nuclear':9, 'Geothermal':10, 'Storage':11, 'Other':12, 'Cogeneration':13, 'Petcoke':14, 'Wave and Tidal':15})
df_all["other_fuel2_num"] = df_all["other_fuel2"].map({'NaN':0, 'Hydro':1, 'Solar':2, 'Wind':3, 'Gas':4, 'Coal':5, 'Oil':6, 'Biomass':7, 'Waste':8, 'Nuclear':9, 'Geothermal':10, 'Storage':11, 'Other':12, 'Cogeneration':13, 'Petcoke':14, 'Wave and Tidal':15})
df_all["other_fuel3_num"] = df_all["other_fuel3"].map({'NaN':0, 'Hydro':1, 'Solar':2, 'Wind':3, 'Gas':4, 'Coal':5, 'Oil':6, 'Biomass':7, 'Waste':8, 'Nuclear':9, 'Geothermal':10, 'Storage':11, 'Other':12, 'Cogeneration':13, 'Petcoke':14, 'Wave and Tidal':15})
df_all.set_index('gppd_idnr')
```

```python
# To count the number of countries. There are a total of 164 countries. Also gives the total number of powerplants for countries.
df_all["country"].value_counts()
```

```python
#Split into one dataset containing estimated and one containing time series data
df_time_series = (df_all.dropna(how='all', subset=['generation_gwh_2013', 'generation_gwh_2014', 'generation_gwh_2015', 'generation_gwh_2016', 'generation_gwh_2017'])
                      .set_index('gppd_idnr'))


df_time_series_complete = df_time_series.dropna(how='any', subset=['generation_gwh_2013', 'generation_gwh_2014', 'generation_gwh_2015', 'generation_gwh_2016', 'generation_gwh_2017'])

df_time_series_incomplete = df_time_series[df_time_series[['generation_gwh_2013', 'generation_gwh_2014', 'generation_gwh_2015', 'generation_gwh_2016', 'generation_gwh_2017']].isnull().any(axis=1)]

df_estimate = (df_all.dropna(how='any', subset=['estimated_generation_gwh'])
                   .set_index('gppd_idnr'))
```

```python
#We must have some duplicates
len(df_time_series_incomplete)+len(df_estimate)+len(df_time_series_complete)
```

```python
#Some rows have both estimate and incomplete time series:
s1 = pd.merge(df_time_series_incomplete, df_estimate, how='inner')
s1
```
